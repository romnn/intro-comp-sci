```python
import numpy as np
import scipy as sp
```

```python
def sin(x):
    return np.sin(x)
```

```python
xdata = np.linspace(0,20, 200)
ydata = sin(xdata)
```

```python
from matplotlib import pyplot as plt
```

```python
#SIMPLE FUNCTION 1
plt.plot(xdata,ydata)
plt.show()
```

```python
ydata2 = np.sin(2*xdata)
```

```python
#SIMPLE FUNCTION 2


plt.plot(xdata,ydata2)
plt.show()
```

```python
# MIXED FUNCTION

plt.plot(xdata,(ydata+ydata2)/2)
plt.show()
```

```python
from scipy import fftpack

weird_one = (ydata+ydata2)/2
```

```python

fig = plt.figure(figsize=(20,20))

# apply fourier analysis
fourier = fftpack.fft(ydata)

# get the frequencies for which it analyses
freqs = fftpack.fftfreq(len(ydata))

plt.plot(freqs, fourier.real)


# apply fourier analysis
fourier = fftpack.fft(ydata2)
# get the frequencies for which it analyses
freqs = fftpack.fftfreq(len(ydata2))


plt.plot(freqs, fourier.real)
plt.title("Analysis of the two simple functions.")

# apply fourier analysis
fourier = fftpack.fft(weird_one)
# get the frequencies for which it analyses
freqs = fftpack.fftfreq(len(weird_one))

plt.show()
fig = plt.figure(figsize=(20,20))

plt.plot(freqs, fourier.real, c='green')
plt.title("Analysis of the mixed function. Note that it finds exactly the same peaks as the figure above (but with half the magnitude)")
plt.show()
```

```python
fig = plt.figure(figsize=(20,20))

# fourier analysis + get frequencies
fourier = fftpack.fft(ydata)
freqs = fftpack.fftfreq(len(ydata))

# sort frequencies & fourier together
freqs, fourier = zip(*sorted(zip(freqs, fourier)))
# index such to plot only positive frequencies
i_min = freqs.index(0)
i_max = freqs.index(0.4)


# plot only indices of positive frequencies
plt.plot(freqs[i_min:i_max], np.abs(np.real(fourier))[i_min:i_max])
plt.title("Fourier analysis for the simple functions. Only the positive frequencies plotted.")

fourier = fftpack.fft(ydata2)
freqs = fftpack.fftfreq(len(ydata2))


# sort frequencies & fourier together
freqs, fourier = zip(*sorted(zip(freqs, fourier)))
# index such to plot only positive frequencies
i_min = freqs.index(0)
i_max = freqs.index(0.4)

# plot only indices of positive frequencies
plt.plot(freqs[i_min:i_max], np.abs(np.real(fourier))[i_min:i_max])

fourier = fftpack.fft(weird_one)
freqs = fftpack.fftfreq(len(weird_one))
plt.show()
fig = plt.figure(figsize=(20,20))

# sort frequencies & fourier together
freqs, fourier = zip(*sorted(zip(freqs, fourier)))
# index such to plot only positive frequencies
i_min = freqs.index(0)
i_max = freqs.index(0.4)

# plot only indices of positive frequencies
plt.plot(freqs[i_min:i_max], np.abs(np.real(fourier))[i_min:i_max], c='green')
plt.title("Positive frequencies of the mixed function")
plt.show()
```
